## Analisis Lanjutan *(RFM Analysis)*

**Tujuan Analisis:**
- Dilakukannya analisis ini untuk menunjukkan pelanggan terbaik berdasarkan parameter RFM (Recency, Frequency, Monetary). Hasil analisis RFM ini akan digunakan untuk menjawab pertanyaan nomor 4, 5, dan 6.

```python
# Membuat DataFrame baru untuk menampung informasi parameter RFM
rfm_df = customers_in_orders_df.groupby(by='customer_id', as_index=False).agg({
    'order_purchase_timestamp': 'max',
    'order_id': 'nunique',
    'price': 'sum'
})
rfm_df.columns = ['customer_id', 'max_order_timestamp', 'frequency', 'monetary']
```

```python
# Menghitung kapan terakhir pelanggan melakukan transaksi (hari)
rfm_df['max_order_timestamp'] = rfm_df['max_order_timestamp'].dt.date
recent_date = orders_df['order_purchase_timestamp'].dt.date.max()
rfm_df['recency'] = rfm_df['max_order_timestamp'].apply(lambda x: (recent_date - x).days)

rfm_df.drop('max_order_timestamp', axis=1, inplace=True)
rfm_df.head()
```

```python
# Membuat visualisasi dari analisis RFM
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(13, 16))

colors = ['#3498db', '#3498db', '#3498db', '#3498db', '#3498db']

sns.barplot(y="customer_id", x="recency", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (days)", loc="center", fontsize=15)
ax[0].tick_params(axis ='y', labelsize=15)

sns.barplot(y="customer_id", x="frequency", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=15)

sns.barplot(y="customer_id", x="monetary", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=15)
ax[2].tick_params(axis='y', labelsize=15)

plt.suptitle("Pelanggan Terbaik Menurut Parameter RFM (customer_id)", fontsize=20, x=0.35)
plt.subplots_adjust(top=0.92)
plt.show()
```

**Insight:**
- Pada grafik **By Recency (days)**, pelanggan dengan ID **4b7decb9b58e2569548b8b4c8e20e8d7** adalah pelanggan yang paling baru melakukan transaksi.
- Pada grafik **By Frequency**, pelanggan dengan ID **00012a2ce6f8dcda20d059ce98491703** memiliki nilai Frequency tertinggi, yang berarti pelanggan ini melakukan pembelian paling sering dibanding pelanggan lainnya.
- Pada grafik **By Monetary**, pelanggan dengan ID **1617b1357756262bfa56ab541c47bc16** memberikan pendapatan terbesar dibanding pelanggan lainnya.
